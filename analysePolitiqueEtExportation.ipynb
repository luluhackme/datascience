```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/refs/heads/master/2011_us_ag_exports.csv')
```

```python
fig, ax = plt.subplots()

ax.pie(df['beef'], labels=df['state'], radius=3)
ax.set_title('% export of beef by states')

plt.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/refs/heads/master/european_leaders.csv')
```

```python
df[(df['country_name'] == 'France') & (df['party_name_english'] == 'Socialist Party')]
```

```python
fig, ax = plt.subplots()

ax.plot(df[(df['country_name'] == 'France') & (df['party_name_english'] == 'Socialist Party')]['median_x'], df[(df['country_name'] == 'France') & (df['party_name_english'] == 'Socialist Party')]['size'])
ax.set_xlabel('Date')
ax.set_ylabel('Nombre d\'adhrents')
ax.set_title('Evolution du Parti Socialiste (2002-2017)')

plt.show()
```
